# About The Dataset
* **Age** : Age of the patient
* **Sex** : Sex of the patient (1 = male, 0 = female)
* **exang**: exercise induced angina (1 = yes; 0 = no)
* **ca**: number of major vessels (0-3)
* **cp** : Chest Pain type chest pain type
    * Value 1: typical angina
    * Value 2: atypical angina
    * Value 3: non-anginal pain
    * Value 4: asymptomatic
* **trtbps** : resting blood pressure (in mm Hg)
* **chol** : cholestoral in mg/dl fetched via BMI sensor
* **fbs** : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* **rest_ecg** : resting electrocardiographic results
    * Value 0: normal
    * Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    * Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
* **thalach** : maximum heart rate achieved
* **target** : 0 = less chance of heart attack, 1 = more chance of heart attack**

# Data Preprocessing

***Split the data into X, y datasets***

```python
X = df.iloc[:,:-1]
y = df["output"]
```

***Standardizing the data***

```python
ss = StandardScaler()
X = ss.fit_transform(X)
```

***Train-Test split***

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Machine Learning Models

```python
models = pd.DataFrame(columns=["Model","Accuracy Score","Confusion Matrix", "Classification Report"])
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
y_pred = log_reg.predict(x_test)
score = accuracy_score(y_test, y_pred)
log_reg_cm = confusion_matrix(y_test, y_pred)
log_reg_cr = classification_report(y_test, y_pred)
print("LogisticRegression: ", score)

new_row = {"Model": "LogisticRegression", "Accuracy Score": score, "Confusion Matrix": log_reg_cm,"Classification Report":log_reg_cr}
models = models._append(new_row, ignore_index=True)
```

```python
GNB = GaussianNB()
GNB.fit(x_train, y_train)
y_pred = GNB.predict(x_test)
score = accuracy_score(y_test, y_pred)
GNB_cm = confusion_matrix(y_test, y_pred)
GNB_cr = classification_report(y_test, y_pred)
print("GaussianNB: ", score)

new_row = {"Model": "GaussianNB", "Accuracy Score": score, "Confusion Matrix": GNB_cm,"Classification Report":GNB_cr}
models = models._append(new_row, ignore_index=True)
```

```python
BNB = BernoulliNB()
BNB.fit(x_train, y_train)
y_pred = BNB.predict(x_test)
score = accuracy_score(y_test, y_pred)
BNB_cm = confusion_matrix(y_test, y_pred)
BNB_cr = classification_report(y_test, y_pred)
print("BernoulliNB: ", score)

new_row = {"Model": "BernoulliNB", "Accuracy Score": score, "Confusion Matrix": BNB_cm,"Classification Report":BNB_cr}
models = models._append(new_row, ignore_index=True)
```

```python
svm = SVC(random_state=0)
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
score = accuracy_score(y_test, y_pred)
svm_cm = confusion_matrix(y_test, y_pred)
svm_cr = classification_report(y_test, y_pred)
print("SVC: ", score)

new_row = {"Model": "SVC", "Accuracy Score": score, "Confusion Matrix": svm_cm,"Classification Report":svm_cr}
models = models._append(new_row, ignore_index=True)
```

```python
rf = RandomForestClassifier(n_estimators=1000, max_depth=5, random_state=42)
rf.fit(x_train, y_train)
predictions = rf.predict(x_test)
score = accuracy_score(y_test, predictions)
rf_cm = confusion_matrix(y_test, y_pred)
rf_cr = classification_report(y_test, y_pred)
print("RandomForestClassifier: ", score)

new_row = {"Model": "RandomForestClassifier", "Accuracy Score": score, "Confusion Matrix": rf_cm,"Classification Report":rf_cr}
models = models._append(new_row, ignore_index=True)
```

```python
xgb = XGBClassifier()
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
score = accuracy_score(y_test, y_pred)
xgb_cm = confusion_matrix(y_test, y_pred)
xgb_cr = classification_report(y_test, y_pred)
print("XGBClassifier: ", score)

new_row = {"Model": "XGBClassifier", "Accuracy Score": score, "Confusion Matrix": xgb_cm,"Classification Report":xgb_cr}
models = models._append(new_row, ignore_index=True)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
score = accuracy_score(y_test, y_pred)
print("KNeighborsClassifier: ", score)
```

***Optimizing the KNN Model***

```python
score_list=[]

for i in range(1,20):
    knn2 = KNeighborsClassifier(n_neighbors=i)
    knn2.fit(x_train, y_train)
    y_pred = knn2.predict(x_test)
    score_list.append(accuracy_score(y_test, y_pred))
```

```python
plt.figure(figsize =(10, 6))
plt.plot(range(1, 20), score_list, marker ='o', markerfacecolor ='red', markersize = 10)
  
plt.title('Score vs K Value', size=15)
plt.xlabel('K value')
plt.ylabel('Score')
```

***The K value that we get the highest accuracy score with is 17, so we are setting the "n_neighbors" parameter to 17***

```python
knn3 = KNeighborsClassifier(n_neighbors=17)
knn3.fit(x_train, y_train)
y_pred = knn3.predict(x_test)
score = accuracy_score(y_test, y_pred)
knn3_cm = confusion_matrix(y_test, y_pred)
knn3_cr = classification_report(y_test, y_pred)
print("KNeighborsClassifier: ", score)

new_row = {"Model": "KNeighborsClassifier", "Accuracy Score": score, "Confusion Matrix": knn3_cm,"Classification Report":knn3_cr}
models = models._append(new_row, ignore_index=True)
```
